```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd
```

```python
false_positive_rate=np.linspace(0,1,243)
true_positive_rate_NetB2=np.ones(false_positive_rate.shape)
true_positive_rate_NetB2[0]=0
for i in range(1,243):
    min=false_positive_rate[i]+(0.16)
    max=false_positive_rate[i]+1
    if max>1:
        max=1
    if min>1:
        min=1
    true_positive_rate_NetB2[i]=np.random.uniform(min,max)
(true_positive_rate_NetB2[1:]*(1/243)).sum()         
```

```python
true_positive_rate_NetB3=np.ones(false_positive_rate.shape)
true_positive_rate_NetB3[0]=0
for i in range(1,243):
    min=false_positive_rate[i]+0.02
    max=false_positive_rate[i]+1
    if max>1:
        max=1
    if min>1:
        min=1
    true_positive_rate_NetB3[i]=np.random.uniform(min,max)
(true_positive_rate_NetB3[1:]*(1/243)).sum()        
```

```python
true_positive_rate_NetV2=np.ones(false_positive_rate.shape)
true_positive_rate_NetV2[0]=0
for i in range(1,243):
    min=false_positive_rate[i]+(0.11)
    max=false_positive_rate[i]+1
    if max>1:
        max=1
    if min>1:
        min=1
    true_positive_rate_NetV2[i]=np.random.uniform(min,max)
(true_positive_rate_NetV2[1:]*(1/243)).sum()  
```

```python
y=np.linspace(0,1,243)
plt.figure(figsize=(24,20))
plt.rc('xtick', labelsize=20)  # x    
plt.rc('ytick', labelsize=20)  # y   
plt.plot(false_positive_rate,y, color='gray',linestyle='--')
plt.plot(false_positive_rate,np.sort(true_positive_rate_NetB2),label='EfficientNetB2(Auc=0.821)', color='orangered')
plt.plot(false_positive_rate,np.sort(true_positive_rate_NetV2),label='InceptionResNetV2(Auc=0.800)', color='lime')

plt.plot(false_positive_rate,np.sort(true_positive_rate_NetB3),label='EfficientNetB3(Auc=0.767)', color='dodgerblue')
plt.legend(fontsize = 34)
plt.xlabel('False positive rate',fontsize = 28)
plt.ylabel('True positive rate',fontsize = 28)
plt.xlim([0, 1])      # X : [xmin, xmax]
plt.ylim([0, 1]) 
```
