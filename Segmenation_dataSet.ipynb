```python
import numpy as np
import slideio
import cv2
import json
import matplotlib.path as mpltPath
import os
import warnings
import tifffile
from skimage import exposure
import xml.etree.ElementTree as ET
import pandas as pd
from PIL import Image
from math import sqrt
import matplotlib.pyplot as plt
from glob import glob
from copy import copy
from tqdm.notebook import tqdm
import warnings
```

```python
svs_list=glob('../../data/OriginalData/svs/*.svs')
xml_list = [f.replace('.svs', '.xml') for f in svs_list]
xml_list = [f.replace('/svs', '/xml') for f in xml_list]
image_list=[f.replace('../../data/OriginalData/svs/', '../../data/CycleGANData/wholeslide/image/') for f in svs_list]
image_list=[f.replace('.svs', '.tiff') for f in image_list]
mask_list=[f.replace('../../data/OriginalData/svs/', '../../data/CycleGANData/wholeslide/mask/') for f in svs_list]
mask_list=[f.replace('.svs', '.tiff') for f in mask_list]
```

```python
size=512
for i in range(len(svs_list)):
    try:
        file_extension_index=svs_list[i].find('.svs')
        fileName=svs_list[i][:file_extension_index]
        slideName=svs_list[i][:9]
        slide = slideio.open_slide(svs_list[i], "SVS")
        scene = slide.get_scene(0)
        svsWidth = scene.rect[2]
        svsHeight = scene.rect[3]
        widthRatio=size/svsWidth
        heightRatio=size/svsHeight 
        imageCount=0
        magnification=80
        tile_size=256
        if(svsHeight>svsWidth):
            RatioMax=heightRatio
            RatioIndex="H"
        else:
            RatioMax=widthRatio
            RatioIndex="W"
        xml_open=xml_list[i]
        polygon_xml = ET.parse(xml_open)
        root = polygon_xml.getroot()
        root=list(root)[0]
        root=list(root)
        imageCount=0
        mask1 = np.zeros((size,size),dtype = np.uint8)
        dstImg = np.ones((size,size,3),dtype = np.uint8)*255
        path=[]
        dstWidth=int(svsWidth*RatioMax)
        dstHeight=int(svsHeight*RatioMax)
        for annotation_count in range(len(root)):
            root1 =list(root[annotation_count])[0]
            polygon=[]
            for polygon_count in range(len(root1)): 
                polygon.append([int(float(root1[polygon_count].attrib['X'])),int(float(root1[polygon_count].attrib['Y']))])
            polygon=np.array(polygon)
            if(RatioIndex=='H'):
                polygon[:,0]=(size-dstWidth)/2+polygon[:,0]*RatioMax
                polygon[:,1]=polygon[:,1]*RatioMax
            else:
                polygon[:,0]=polygon[:,0]*RatioMax
                polygon[:,1]=(size-dstHeight)/2+polygon[:,1]*RatioMax
            path.append(np.array(polygon))
        cv2.fillPoly(mask1,path,255)
        thumbnailImg=scene.read_block((0,0,svsWidth,svsHeight),size=(dstWidth,dstHeight ))
        if(RatioIndex=='H'):
            dstImg[:,int((size-dstWidth)/2):int((size+dstWidth)/2)]=thumbnailImg
        else:
            dstImg[int((size-dstHeight)/2):int((size+dstHeight)/2),:]=thumbnailImg            
        dstImg=cv2.cvtColor(dstImg,cv2.COLOR_BGR2RGB)

        cv2.imwrite(image_list[i],dstImg)
        cv2.imwrite(mask_list[i],mask1)
    except:
        file_extension_index=svs_list[i].find('.svs')
        fileName=svs_list[i][:file_extension_index]
        print(fileName)
```
