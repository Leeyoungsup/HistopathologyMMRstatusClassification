```python
from PIL import Image
import numpy as np 
import matplotlib.pyplot as plt
```

```python
image=np.array(Image.open('../../data/BLOP-AJ-00049-S-TP-03_2.jpg'))
def plot_color_statistics(image_path):
    # 이미지 불러오기
    image = plt.imread(image_path)

    # 이미지의 높이와 너비
    height, width, _ = image.shape

    # 이미지의 RGB 값을 2D 배열로 변환
    rgb_values = image.reshape((height * width, 3))

    # RGB 평균값과 표준편차 계산
    average_rgb = np.mean(rgb_values, axis=0)
    std_rgb = np.std(rgb_values, axis=0)
    plt.figure(figsize=(10,10))
    # 막대그래프 플로팅
    plt.bar(range(3), average_rgb, yerr=std_rgb, capsize=5, color=['red', 'green', 'blue'])
    plt.title('Average RGB Values with Standard Deviation',fontsize=15)
    plt.xlabel('Color Channels',fontsize=15)
    plt.ylabel('Value',fontsize=15)
    plt.xticks(range(3), ['Red', 'Green', 'Blue'],fontsize=15)
    plt.show()
    print(average_rgb,std_rgb)
def plot_combined_color_distribution(image_path):
    # 이미지 로드
    image = Image.open(image_path)

    # 이미지를 NumPy 배열로 변환
    image_array = np.array(image)

    # 이미지의 높이, 너비 및 채널 수 가져오기
    height, width, channels = image_array.shape

    # 이미지 픽셀을 1차원 배열로 변환
    flattened_image = image_array.reshape((height * width, channels))

    # 각 채널에 대한 히스토그램 계산
    hist_red, bins_red = np.histogram(flattened_image[:, 0], bins=256, range=[0, 256])
    hist_green, bins_green = np.histogram(flattened_image[:, 1], bins=256, range=[0, 256])
    hist_blue, bins_blue = np.histogram(flattened_image[:, 2], bins=256, range=[0, 256])

    # 그래프 그리기
    plt.figure(figsize=(10,10))
    plt.legend(fontsize = 20)
    # 각 채널의 히스토그램을 하나의 막대그래프로 표시
    plt.plot(hist_red, color='red', label='Red Channel', alpha=1)
    plt.plot(hist_green, color='green', label='Green Channel', alpha=1)
    plt.plot(hist_blue, color='blue', label='Blue Channel', alpha=1)
    

    plt.title('Combined Color Distribution',fontsize=15)
    plt.xlabel('Pixel Value',fontsize=15)
    plt.ylabel('Frequency',fontsize=15)
    plt.legend(fontsize='large')
    plt.show()
plot_color_statistics('../../data/BLOP-AJ-00049-S-TP-03_2.jpg')
plot_combined_color_distribution('../../data/BLOP-AJ-00049-S-TP-03_2.jpg')
```
